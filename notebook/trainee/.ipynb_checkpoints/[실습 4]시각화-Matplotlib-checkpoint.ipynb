```python
import io
import requests

def get_csv(url):
    content=requests.get(url).content
    return pd.read_csv(io.StringIO(content.decode('utf-8')))

url = 'https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv'

housing = get_csv(url)
housing.head()
```
